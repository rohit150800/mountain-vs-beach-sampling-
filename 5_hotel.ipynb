```python
#encoding
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['room_type_reserved']=label.fit_transform(df['room_type_reserved'])
df['market_segment_type']=label.fit_transform(df['market_segment_type'])
```

```python
inputs=df.iloc[:,:-1].values
outputs=df.iloc[:,-1].values
```

```python
# Spliting
from sklearn.model_selection import train_test_split
in_train,in_test,out_train,out_test=train_test_split(inputs,outputs,test_size=.30,random_state=42)
```

```python
# normalisation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(in_train)
in_train=scaler.transform(in_train)
in_test=scaler.transform(in_test)
```

```python
# MOdel Creation 
# COMBINE THREE MODELS

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB  #After normalisation there is negative value in the input  , For negative values , we use Bernoulli, WE cant use the first 2 classifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
knn=KNeighborsClassifier(n_neighbors=7)
nav=BernoulliNB()
sup=SVC()
lst=[knn,nav,sup]

```

```python
for i in lst:
    print("Model Name:'",i,"'")
    print('*'*100)
    i.fit(in_train,out_train)
    out_pred=i.predict(in_test)
    print("PERFORMANCE EVALUTUION:")
  
    print(confusion_matrix(out_test,out_pred))
    print("accuracy score:",accuracy_score(out_test,out_pred))
    print(classification_report(out_test,out_pred))
    print()
    print()
    
```
